# Convert mp4 into gif

## Lightweight GIF

- 480px width
- 12 FPS

```
from google.colab import files
uploaded = files.upload()

from moviepy.editor import VideoFileClip

# Change this to your uploaded file name
input_path = list(uploaded.keys())[0]
output_path = "output.gif"

clip = VideoFileClip(input_path)
clip_resized = clip.resize(width=480)
clip_resized.write_gif(output_path, fps=12)

files.download(output_path)
```

## Ultra-Lightweight High-Quality GIF Code

- 400px width → biggest impact on size, still sharp

- 12 FPS → smooth enough but much smaller

- ffmpeg optimization → reduces size while keeping detail

- Full 256 colors → keeps quality high

```
from google.colab import files
uploaded = files.upload()

from moviepy.editor import VideoFileClip

# Get uploaded filename
input_path = list(uploaded.keys())[0]
output_path = "lightweight_high_quality.gif"

# Load video
clip = VideoFileClip(input_path)

# ✅ Resize to 400px width → big size reduction, still looks sharp
clip_resized = clip.resize(width=400)

# ✅ Export optimized GIF
clip_resized.write_gif(
    output_path,
    fps=12,
    program='ffmpeg',
    opt='wu'      # wu = optimized, smaller output
)

# Download the GIF
files.download(output_path)
```
